## 11. Folium
- 지도 시각화 라이브러리

```python
import folium
```

```python
m = folium.Map(location=[37.61119807384533, 126.91731049700418],
                zoom_start=18,
                #tiles="Cartodb Positron"
                )
# 기본 마커
folium.Marker([37.61119807384533, 126.91731049700418], 
                popup="Subway",
                tooltip="구산역",
                icon=folium.Icon(color="blue", icon="fa-solid fa-train", prefix="fa-solid")
                ).add_to(m)

# 원형 마커
folium.CircleMarker(
    [37.61020411781574, 126.9133136519163],
    radius=100,
    color = "#adcdff",
    fill_color = "#1c73ff",
    popup= "CircleMarker",
    tooltip="tooltip"
).add_to(m)

# 클릭한 곳에 마커 추가
m.add_child(folium.ClickForMarker(popup="내가 클릭한 곳"))

# 클릭한 속의 위도와 경도 표시
m.add_child(folium.LatLngPopup())

m
```

```python
# 실습. 강의장 주변 장소 3개 표시하기
m = folium.Map(location=[37.610320465, 126.912954327],
                zoom_start=18,
                #tiles="Cartodb Positron"
                )
# 기본 마커
folium.Marker([37.610320465, 126.912954327], 
                popup="Here",
                tooltip="교육장",
                icon=folium.Icon(color="pink", icon="info-sign")
                ).add_to(m)

# 원형 마커
folium.CircleMarker(
    [37.610320465, 126.912954327],
    radius=100,
    color = "#adcdff",
    fill_color = "#1c73ff",
    popup= "CircleMarker",
    tooltip="tooltip"
).add_to(m)

# 클릭한 곳에 마커 추가
m.add_child(folium.ClickForMarker(popup="강의장 주변 장소"))

# 클릭한 속의 위도와 경도 표시
m.add_child(folium.LatLngPopup())

m
```

```python
# 실습. 택시 승차대 위치 표시하기
# “택시승차대 현황.csv”의 데이터로 택시 승차대 위치 표시하기!

import pandas as pd

df = pd.read_csv("dataset/택시승차대 현황.csv", encoding="CP949")

taxi = df.loc[:, ["위치명", "위도", "경도"]]

latitude = taxi["위도"].mean()
longitude = taxi["경도"].mean()

m = folium.Map([latitude, longitude], zoom_start=13)

for idx, row in taxi.iterrows():            # iterrows : 한행, 한행을 ~
    #print(idx, row["위치명"], row["위도"], row["경도"])
    folium.Marker(
        [row["위도"], row["경도"]],
        tooltip = row["위치명"],
        icon=folium.Icon(color="orange", icon="fa-solid fa-taxi", prefix="fa-solid")
    ).add_to(m)

m
```
