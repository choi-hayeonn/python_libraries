```python
# ÏÇ¨Ïö© Í∞ÄÎä•Ìïú Ìè∞Ìä∏ ÌôïÏù∏
import matplotlib.font_manager as fm

for f in fm.fontManager.ttflist : 
    print(f.name)
```

```python
# Seaborn Ïä§ÌÉÄÏùº ÏÑ§Ï†ï
# Î∞©Î≤ï 1.
sns.set_style("ticks")  # Ïä§ÌÉÄÏùº ÌÖåÎßà ÏÑ§Ï†ï # ÏÇ¨Ïö© Í∞ÄÎä•Ìïú ÌÖåÎßà : ‚Äúwhite‚Äú, ‚Äúdark‚Äú, ‚Äúwhitegrid‚Äú, ‚Äúdarkgrid‚Äú, ‚Äúticks‚Äù
sns.set_context("notebook") # Î¨∏Îß•Ïóê Îî∞Îùº Ïä§ÌÉÄÏùº ÌÅ¨Í∏∞ Ï°∞Ï†ï # ÏÇ¨Ïö© Í∞ÄÎä•Ìïú context : ‚Äúpaper‚Äú - ÎÖºÎ¨∏/Ï∂úÌåêÏö©, ‚Äúnotebook‚Äú - Jupyter Í∏∞Î≥∏ (Í∏∞Î≥∏Í∞í), ‚Äútalk‚Äú ‚Äì ÌîÑÎ†àÏ††ÌÖåÏù¥ÏÖòÏö©, ‚Äúposter‚Äú - Ìè¨Ïä§ÌÑ∞ Îì± ÎåÄÌòï ÎîîÏä§ÌîåÎ†àÏù¥Ïö©
sns.set_palette("pastel6")   # ÏÉâÏÉÅ ÏßÄÏ†ï
# ÌïúÍ∏Ä Íπ®Ïßê Î∞©ÏßÄ
plt.rcParams['font.family'] = 'Malgun Gothic'  # 'ÎßëÏùÄ Í≥†Îîï'Ïù¥ ÏÑ§ÏπòÎêòÏñ¥ ÏûàÏùÑ Í≤ΩÏö∞
plt.rcParams['axes.unicode_minus'] = False     # ÎßàÏù¥ÎÑàÏä§(-) Î∂ÄÌò∏ Íπ®Ïßê Î∞©ÏßÄ

# Î∞©Î≤ï 2.
sns.set_theme(
    style = "darkgrid",
    rc = {
        "font.family" : "Malgun Gothic",
        "axes.unicode_minus" : False
    }
)
```

```python
# Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞ Í∑∏ÎûòÌîÑ
# countplot
palette = sns.color_palette("pastel")  # ÏÉâÏÉÅ Îî∞Î°ú(Î™ÖÏãúÏ†ÅÏúºÎ°ú) ÏßÄÏ†ï
tips = sns.load_dataset("tips")
sns.countplot(data = tips, x = "day", hue = "sex", palette=palette, alpha=0.7)  # hue: Îã§Î•∏ Î≤îÏ£ºÎ°ú Í∑∏Î£πÌôîÌïòÏó¨ Îã§Ï§ë ÏãúÍ∞ÅÌôî Í∞ÄÎä•
plt.title("ÏöîÏùºÎ≥Ñ Î∞©Î¨∏Ïûê Ïàò")
plt.grid(linestyle=":")
plt.show()
```

```python
# barplot
sns.barplot(data=tips, x="day", y="tip", palette=palette, alpha=0.6)

plt.title("ÏöîÏùºÎ≥Ñ ÌèâÍ∑† ÌåÅ Í∏àÏï°")
plt.xticks(rotation=45)
plt.show()
```

```python
# displot()
# ÌûàÏä§ÌÜ†Í∑∏Îû® + Ïª§ÎÑê Î∞ÄÎèÑ(kde)
# kde=TrueÎ°ú ÌïòÎ©¥ ked Í∑∏ÎûòÌîÑ Í∑∏Î†§Ï§å
sns.displot(data=tips, x="total_bill", kde=True, bins=20, palette=palette, color="#60BF79")
plt.title("Total bill")
plt.show()
```

```python
# scatter()
# ÏÇ∞Ï†êÎèÑ
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", palette=palette)
plt.title("Total billÍ≥º tipÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ")
plt.show()
```

```python
# heatmap()
# ÏÉÅÍ¥ÄÍ≥ÑÏàò ÌñâÎ†¨Ïùò Í∑∏ÎûòÌîÑ
penguins = sns.load_dataset("penguins")    # bill : Î∂ÄÎ¶¨  # flipper : ÎÇ†Í∞ú
penguins
```

```python
# ÏÉÅÍ¥ÄÍ≥ÑÏàò ÌñâÎ†¨
corr= penguins.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="YlGn", alpha=0.5)
plt.show()
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind='reg')
plt.show()

# Ï§ëÏã¨: ÏÇ∞Ï†êÎèÑ + ÌöåÍ∑ÄÏÑ† (kind='reg')
# Ï£ºÎ≥Ä: x/y Î≥ÄÏàò Í∞ÅÍ∞ÅÏùò Î∂ÑÌè¨ (ÌûàÏä§ÌÜ†Í∑∏Îû®)
# ÏÉÅÍ¥ÄÏÑ±Í≥º Î∂ÑÌè¨Î•º ÎèôÏãúÏóê ÌôïÏù∏ Í∞ÄÎä•
```

```python
# jointplot()
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="hex", alpha=0.7, color="#60BF79")
plt.show()
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="kde", alpha=0.7, color="#f086a5")
plt.show()
```
