### 조건 필터링

```python
import pandas as pd
import numpy as np
```

```python
data = {
    '이름': ['홍길동', '김철수', '이영희', '박민수'],
    '점수': [90, 70, 95, 95],
    '반': [2, 1, 1, 2]
}
df = pd.DataFrame(data)
df
```

```python
mask = df["점수"] >= 80
filtered = df[mask]
filtered
```

```python
# 여러조건 결합
# and : &
# or : |
# not : ~

# & 예시
df[(df["반"] == 1) & (df["점수"] >= 80)]
```

```python
# | 예시
mask = (df["반"] == 2) | (df["점수"] >= 90)
df[mask]
```

```python
# isin()
# 여러 값 중에 하나에 해당하는지 여부를 판단할 때 사용

# (df["이름"] == "홍길동") | (df["이름"] == "박민수") | (df["이름"] == "이영희") 
# 위와 같이 복잡한 걸 아래와 같이 간단하게 표현 가능.
mask = df["이름"].isin(["홍길동", "박민수"])
df[mask]
```

```python
# not 사용 예시
mask = ~df["이름"].isin(["홍길동", "박민수"])
df[mask]
```

```python
# reset index
# 조건 필터링, 정렬 등을 통해 인덱스가 변경됐을 시
# reset_index를 사용해서 인덱스를 초기화

mask = (df["반"] == 2) | (df["점수"] >= 90)
df2 = df[mask]
print(df2)
print()
# 인덱스 리셋
df3 = df2.reset_index()
print(df3)
print()
# 기존 인덱스 삭제
df4 = df2.reset_index(drop=True)
print(df4)
```

### groupby
- 항상 집계 함수와 함께 사용

```python
df = pd.DataFrame({
    'team': ['A', 'A', 'B', 'B', 'B', 'C'],
    'name': ['Kim', 'Lee', 'Park', 'Choi', 'Jung', 'Han'],
    'score': [90, 85, 80, 70, 95, 88]
})
df
```

```python
# 단일 컬럼 기준 그룹화
grouped = df.groupby("team")

# 집계함수 적용

# 합계
result_sum = grouped["score"].sum()
print(result_sum)

# 평균
result_mean = grouped["score"].mean()
print(result_mean)

# 갯수
result_count = grouped["score"].count()
print(result_count)
```

```python
df2 = pd.DataFrame({
    'team': ['A', 'A', 'B', 'B', 'B', 'C'],
    'gender': ['M', 'F', 'F', 'M', 'M', 'F'],
    'score': [90, 85, 80, 70, 95, 88], 
    'age' : [21,22,23,25,20,27]
})
df2
```

```python
# 여러 컬럼 기준 그룹화
result = df2.groupby(["team", "gender"])["score"].mean()
result
```

```python
# as_index=False 옵션
# 그룹라벨이 인덱스로 설정됨
# result = df2.groupby("team")["score"].sum()
result = df2.groupby("team", as_index=False)["score"].sum()
result
```

```python
# 여러 집계함수를 한번에 적용 : agg()
result = df2.groupby("team", as_index=False)["score"].agg(["sum", "mean", "count"])
result
```

```python
# 그룹별로 여러 컬럼에 다른 집계 함수 적용
result = df2.groupby("team").agg({
    "score" : "mean",
    "age" : "max"
})
result
```

```python
# 실습. groupby 연습문제
# 1. 각 학년(grade)별 평균 국어 점수(kor)를 구하세요.
df = pd.DataFrame({
'grade': [1, 2, 1, 2, 1, 3],
'name': ['Kim', 'Lee', 'Park', 'Choi', 'Jung', 'Han'],
'kor': [85, 78, 90, 92, 80, 75]
})
result = df.groupby("grade", as_index=False)["kor"].mean()
result
```

```python
# 2. 아래 DataFrame에서 반(class)별, 과목(subject)별로 시험에 응시한 학생 수(count)와 평균 점수(avg)를 구하세요.
df = pd.DataFrame({
'class': [1, 1, 1, 2, 2, 2],
'subject': ['Math', 'Math', 'Eng', 'Math', 'Eng', 'Eng'],
'score': [80, 90, 85, 70, 95, 90]
})

result = df.groupby(["class", "subject"])["score"].agg(["count", "mean"])
result
```

```python
# 3. 아래 DataFrame에서 지역(region)별 판매자(seller)별로 판매액(sales)의 합계와 최대값을 구하세요.
df = pd.DataFrame({
'region': ['Seoul', 'Seoul', 'Busan', 'Busan', 'Daegu'],
'seller': ['A', 'B', 'A', 'B', 'A'],
'sales': [100, 200, 150, 120, 130]
})

result = df.groupby(["region", "seller"])["sales"].agg(["sum", "max"])
result
```

```python
# 4. 아래 DataFrame에서 팀(team)별, 포지션(position)별로 결측치(NaN)를 포함한 점수(score)의 평균을 구하세요.
df = pd.DataFrame({
'team': ['A', 'A', 'B', 'B', 'A', 'B'],
'position': ['FW', 'DF', 'FW', 'DF', 'DF', 'FW'],
'score': [3, 2, None, 1, 4, 2]
})

result = df.groupby(["team", "position"], as_index=False)["score"].mean()
result
```

```python
# 5. 아래 DataFrame에서 부서(dept)별로 성별(gender)별 인원 수와, 총 연봉(salary) 합계를구하세요.
df = pd.DataFrame({
'dept': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales'],
'gender': ['M', 'F', 'F', 'M', 'F', 'F'],
'salary': [3500, 3200, 4000, 4200, 3000, 3100]
})

# result = df.groupby(["team", "position"], as_index=False)["score"].agg(['count', 'sum'])

result = df.groupby(['dept', 'gender']).agg(
    count=("salary", "count"),
    total_salary= ("salary", "sum")
)
result
```
