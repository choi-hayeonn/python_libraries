## 07. matplotlib
- 파이썬에서 가장 널리 사용되는 데이터 시각화 라이브러리!!

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


print(matplotlib.__version__)

# 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'  # '맑은 고딕'이 설치되어 있을 경우
plt.rcParams['axes.unicode_minus'] = False     # 마이너스(-) 부호 깨짐 방지
```

### 그래프 그리기

```python
x = [1,2,3,4,5]
y = [10,13,17,19,14]

# 선 그래프
plt.plot(x,y)

# 이미지를 파일로 저장
plt.savefig("my_graph.png")

# 콘솔에 이미지 그리기
plt.show()
```

### 그래프의 구성 요소 (1)
- figure : 여러 그래프가 그려지는 최상위 컨테이너
- axes : 실제 그래프가 그려지는 공간
- axis : 그래프의 축

```python
# subplots()
# fig, axs = plt.subplots(행 , 열)
fig, axs = plt.subplots()    # 하나의 figure와 하나의 axes를 만듦.
```

```python
# 1행 2열의 axes 생성
fig, axs = plt.subplots(1,2)
axs[0].plot([1,2,3],[1,2,3])
axs[1].plot([1,2,3],[1,4,9])

plt.show()
```

```python
# 2행 2열의 axes 생성
fig, axs = plt.subplots(2,2)

axs[0,0].plot([1,2,3],[1,2,3])
axs[0,1].plot([1,2,3],[1,4,9])
axs[1,0].plot([1,2,3],[3,2,1])
axs[1,1].plot([1,2,3],[9,4,1])
plt.tight_layout()  # 그랲 간격 자동 조정(참고용)

plt.show()
```

### 그래프의 구성 요소 (2)
- 제목, 레이블, 범례, 그리드

```python
# title - 기본
plt.plot([1,2,3,4,5], [1,4,9,16,7], 
        label="제곱 그래프", linestyle="--", 
        color="y", marker="o")
plt.title("제곱 함수 그래프", color="g", fontsize=15)

# xlabel, ylabel - 기본
plt.xlabel("x 값")
plt.ylabel("y 값(x의 제곱)")

# xtixks, yticks
plt.xticks([1,2,3,4,5],["A","B", "C", "D", "E"])
plt.yticks([1,5,10,15,20])

# 범례(legend)
plt.legend()

# 격자(grid)
plt.grid(linestyle="--")

plt.show()
```

```python
# title - subplot
fig, axs = plt.subplots()

axs.plot([1,2,3], [1,4,9], "bo--",label="제곱 그래프")
axs.set_title("제곱 함수 그래프")

# xlabel, ylabel - subplot
axs.set_xlabel("x 값")
axs.set_ylabel("y 값(x의 제곱)")

# 범례(legend)
axs.legend()

# 격자(grid)
axs.grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2)

fig.suptitle("여러 함수 그래프")
axs[0,0].plot([1,2,3],[1,2,3])
axs[0,0].set_title("1번 그래프")
axs[0,1].plot([1,2,3],[1,4,9])
axs[0,1].set_title("2번 그래프")
axs[1,0].plot([1,2,3],[3,2,1])
axs[1,0].set_title("3번 그래프")
axs[1,1].plot([1,2,3],[9,4,1])
axs[1,1].set_title("4번 그래프")

plt.show()
```

```python
# 하나의 axes에 여러개의 그래프 그리기
x = np.arange(0,10)
y1 = 2*x+1
y2 = 3*x-5

plt.plot(x, y1, marker="o", label="y= 2x + 1", color="c")
plt.plot(x, y2, marker="*", label="y= 3x - 5", color="m")


plt.title("두개의 선 그래프 비교")
plt.xlabel("x 값")
plt.ylabel("y 값")
plt.legend()
plt.grid(linestyle=":", alpha=0.5)

plt.show()
```

```python
# 실습. 선 그래프 그리기
# 아래 데이터를 이용해서 예시와 같이 선 그래프를 그려보세요.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]


plt.plot(months, sales_2019, marker="o", label="2019", color="g", linestyle=":")
plt.plot(months, sales_2020, marker="^", label="2020", color="y", linestyle="--")

plt.title("Monthly Sales Comparison (2019-2020)", color="g", fontsize=15)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(loc="lower right", title="판매연도")
plt.grid(linestyle=":", alpha=0.5)

plt.show()
```

```python
# 막대 그래프
# 수직 막대 그래프 그리기
labels = ["A","B", "C", "D"]
values = [10,24,15,32]
plt.bar(labels, values)
plt.show()
```

```python
# 수평 막대 그래프 그리기
plt.barh(labels, values)
plt.show()
```

```python
# 여러 막대 그래프
labels = ["1분기","2분기", "3분기", "4분기"]
men_means = [20,34,30,35]
women_means = [25,32,34,40]

x = np.arange(len(labels))

fig, ax = plt.subplots()

# 겹치지 않게 그래프 위치 조정
width = 0.4
ax.bar(x - width/2, men_means, width, label="남성")
ax.bar(x + width/2, women_means, width, label="여성")

ax.set_xticks(x, labels)

plt.show()
```

```python
# 실습. 막대 그래프 그리기
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45] 
plt.bar(categories, data, color="c")

plt.title("Bar Chart", fontsize=15)
plt.xlabel("Categories")
plt.ylabel("values")
plt.grid(linestyle=":", alpha=0.7)
plt.xticks(rotation=45) # X축 텍스트 45도 회전
plt.ylim(0,50) # 그래프의 표시 범위를 제한해주는 것

plt.show()
```

```python
# 히스토그램
# 랜덤 데이터 생성
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor="k")

plt.show()
```

```python
# 실습. 히스토그램 그리기
# 임의로 1~6 사이의 주사위 눈 100개를 생성해서 각 숫자가 얼마나 자주 나왔는지 히스토그램을 그려보세요.

np.random.seed(42)  # 난수의 초기값 지정
data = np.random.randint(1, 7, 100) 

# cnt,bins,_ = plt.hist(data, bins=6, edgecolor='k', alpha=0.5, rwidth=0.8)
plt.hist(data, bins=np.arange(1,8)-0.5, edgecolor='k', alpha=0.5, rwidth=0.8)
# plt.hist(data, bins=np.arange(0.5, 7,5), edgecolor='k', alpha=0.5)
x = np.arange(7)
plt.xticks(x)
plt.ylim(0,25)
plt.xlabel('주사위 눈')
plt.ylabel('빈도')
plt.title('주사위 눈 히스토그램')
plt.grid(linestyle=":", alpha=0.5)

plt.show()
```

```python
# 산점도(Scatter Plot)
# x = [5,7,8,10,12]
# y = [20,25,15,30,22]

x = np.linspace(2, 30, 20)
y = np.linspace(10, 40, 20) 
y_noisy = y + np.random.uniform(-7,7,20)    # uniform : 지정한 구간 안에서 실수 난수 생성

plt.scatter(x,y_noisy)
plt.grid(linestyle=":", alpha=0.5)
plt.show()
```

```python
# 다차원 데이터 표현
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = np.random.rand(N) * 1000

plt.scatter(x,y,s=sizes, c=colors, alpha=0.7, cmap="viridis")
plt.colorbar()

plt.show()
```

```python
# 실습. 산점도 그리기
# x값은 0부터 10까지 30개의 균등한 간격으로 구성된 수열을 만들고,
# y값은 y = 3x + 2 + 난수(정규분포, 평균 0, 표준편차 3)로 생성하여 산점도를 그려보세요.

x = np.linspace(0,10,30)
noise = np.random.normal(0,3,30)
y = 3 * x + 2 + noise


plt.scatter(x,y, c = colors, alpha = 0.7, cmap = "Blues") 
plt.grid(linestyle=":", alpha=0.5)

plt.xlabel('x 값')
plt.ylabel('y 값')
plt.ylim(0,35)
plt.title('노이즈가 추가된 산점도')

plt.show()
```

```python
# 파이 차트
labels = {"A", "B", "C", "D"}
sizes = [40,20,25,15]
explode = [0.2, 0.1, 0.1, 0.1]

plt.pie(sizes, labels = labels, autopct="%1.1f%%", shadow=True, explode=explode, startangle=90, 
        colors=["#f0dae3","#d0e4f5", "#f5e7c1",  "#c7f2dc"])
plt.axis("equal")
plt.show()
```

```python
# 실습5. 파이 차트 그리기
# 아래와 데이터를 이용해서 예시와 같이 파이 차트를 그려보세요!

labels = ['Apple', 'Banana', 'Mango', 'Blueberry']
sizes = [15,30,45,10] 
explode = (0,0.1,0,0)

plt.pie(sizes, labels = labels, autopct="%1.1f%%", shadow=True, explode=explode, startangle=90,
        colors=["mistyrose","oldlace", "lemonchiffon",  "lavender"])
plt.axis("equal")
plt.show()
```
